```python
# csv library---> Most of the time we are going to work with csv files so thats why 
# knowledge with csv library functionality matters
```

```python
import csv
```

```python
with open('example.txt',mode='w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['Name', 'Age', 'City'])
    writer.writerow(['Alice', 25, 'New York'])
    writer.writerow(['Bob', 30, 'San Francisco'])
```

```python
with open('example.txt',mode='r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

```python
# Json Library

# When we are going to work with nosql database then data inside the nosql database is going 
# to be there in the form of json format and along with that when we are going to work with
# the datasets as well then again these dataset will be consisting of data in the json format itself
```

```python
# When we get the json data, we get in a json string format so by using json library we can 
# convert our string data into dictionary format for serializing our data
```

```python
import json
```

```python
data={'Name':'Max','Age':25}
json_string=json.dumps(data)
print(json_string)
print(type(json_string))
```

```python
json_dict=json.loads(json_string)
print(json_dict)
print(type(json_dict))
```

```python
# Regular Expression
# Regular Expression is used to retrieve info based on a specific pattern or if the pattern matches
```

```python
import re
```

```python
text='Hello My name is Max and my password is Max@123'
```

```python
pattern=r'\d+'
match=re.search(pattern,text)
print(match.group())
```
