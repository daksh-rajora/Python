# Conditional Statements

## Conditional Statements works like Real Life Scenario where if a certain condition is meeting then a statement of code or a block of code gets executed

```python
# Eligibility for driving on meeting a certain condition 

# Condtional Statements ---> if Statements where if a condition is meeting then a block of code gets executed

Age=int(input("Enter Your Age"))

if Age>=18:
    print('You are eligible for driving')
else:
    print('You are not eligible for driving')
```

# Conditional statements
## if statement
## elif statement
## else statement

```python
# Figure out that who is taller among two friends

Max_height=float(input("Enter your Height Max:"))
Lauren_height=float(input("Enter your Height Lauren:"))

if Max_height>Lauren_height:
    print("Max is taller than Lauren")
else:
    print("Lauren is taller than Max")
```

```python
# Figure out that who is taller among three friends

Max_height=float(input("Enter your Height Max:"))
Lauren_height=float(input("Enter your Height Lauren:"))
Tony_height=float(input("Enter your Height Tony"))

if Max_height>Lauren_height>Tony_height:
    print("Max is taller than among three friends")
elif Lauren_height>Max_height>Tony_height:
    print("Lauren is taller than among three friends")
else:
    print("Tony is taller than among three friends")
```

# Nested Conditional Statement

### Nested Conditonal statements involve if statements inside another if statement

```python
# Compare the height of three friends
Max_height=float(input("Enter your Height Max:"))
Lauren_height=float(input("Enter your Height Lauren:"))
Tony_height=float(input("Enter your Height Tony"))

if Max_height > Lauren_height and Max_height > Tony_height:
    print("Max is taller")
elif Lauren_height > Max_height and Lauren_height > Tony_height:
    print("Lauren is taller")
else:
    print("Tony is taller")
```

```python
# Whether if the current year is a leap year or not

year=int(input("Enter the Year"))

if year%4==0:
    if year%100==0:
        if year%400==0:
            print("Leap Year")
        else:
            print("Not a Leap Year")
    else:
        print('Year is a leap year')
else:
    print('Year is not a leap year')
```
