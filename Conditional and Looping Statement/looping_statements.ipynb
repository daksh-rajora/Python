# Range()
## range() function can be used for working as our iterable

### range() take an input parameter and the first paramenter that the range will be taking as the input will be the initial value that you want to print or you want to operate on and then you will be passing the next input after comma as the parameter as the final value and it can take third parameter as well and that will be the step size and that is the difference between the iterating values

```python
# print the number from 1 to 50
# range(initial_value, final_value + 1, step_size)

for i in range(1, 51,1):
    print(i)
```

### in the range there is never a neccessity for mentioning the step size and similarly if you want to print the numbers or you want to operate on numbers again there is no neccessity for passing the initial value as well if and only if your initial value is 0

```python
# numbers from 0 to 100
for i in range(101):
    print(i)
```

```python
# print numbers from 0 to 100 with gap of 5

for i in range(0,101,5):
    print(i)
```

## By using range() we can represent the numbers in decreamenting manner as well or in decreasing order as well

```python
# print the numbers from 50 to 1 with a gap of 1 value in a decreamenting manner

for i in range(50,0,-1):
    print(i)
```

```python
# print the numbers from 50 to 0 with a gap of 5

for i in range(50,-1,-5):
    print(i)
```

```python
# print numbers from 500 to 0 with a gap of 5

for i in range(500, -1, -5):
    print(i)
```

# While loop

## while loop is being used for iterating over a series of numbers when a condition is True and we have to initialise a value as well while iterating

```python
# by making the use of while loop we need to print the numbers from 0 to 50

i=0
while i<=50:
    print(i)
    i=i+1

```

```python
# we have to print the numbers from 0 to 50 but even numbers only
for i in range(51):
    if i%2==0:
        print(i)
```

```python
# printing even numbers using while loop in between 0 to 50

i=0
while i<=50:
    if i%2==0:
        print(i)
    else:
        pass
    i=i+1
```

```python
# infinite loop --> infinite loop is being used in a way where in the condition is not met or
#  we are not putting a condition for loop to terminate

i=1
while i>0:
    print(i)
    i=i+1

## this will run infinite loop
```

# Loop Control Statements

## Loop Control statements are being used for terminating a loop in between when a condition is met or if you want to have some control over your iterating process or while iterating over the series of numbers

### break---> break is our first loop control statement which will  help you to terminate the loop when a certain condition is met

```python
## print the number from 0 to 50 but once number=18 then terminate the loop or do not print
#  either 18 or any number forward

for i in range(51):
    if i==18:
        break
    else:
        print(i)
```

### continue-->continue is our second loop control statement which helps us in iterating over a series of numbers but helps us with skipping the values in between once we met a condition

```python
## print the numbers from 0 to 50 but only print even values skip odd one
for i in range(51):
    if i%2!=0:
        continue
    else:
        print(i)
```

```python
## print the numbers from 0 to 20 skip 17

for i in range(21):
    if i==17:
        continue
    else:
        print(i)
```

## Pass --> pass statement is being used for doing nothing

```python
## print the numbers from 0 to 20 but once 17 occur do nothing
for i in range(21):
    if i==17:
        pass
    else:
        print(i)
```

# Nested loops

## Nested loops --> it is going to have a loop inside a loop

### f string --> f string will be helping us out with printing the numbers with the variable value present inside the colons so we can directly represent the variable value inside the colons message itself where variable value will be shown inside the {}

```python
# Nested loops can help us out with iterating over a series of numbers as with the
# proper combination of iterating numbers

for i in range(6):
    for j in range(4):
        print(f'i value is :{i} and j value is {j}')
```

```python
## printing the sum of first n natural numbers where natural number ---> 1,2,3,.....n
n=int(input('Enter the natural number'))
sum=0
for i in range(n+1):
    sum+=i
    print(sum)
```

```python
## print prime number from 0 to 100
for num in range(2,101):
    for j in range(2,num):
        if num%j==0:
            break
        print(num)
```
