# Range()
## range() function can be used for working as our iterable

### range() take an input parameter and the first paramenter that the range will be taking as the input will be the initial value that you want to print or you want to operate on and then you will be passing the next input after comma as the parameter as the final value and it can take third parameter as well and that will be the step size and that is the difference between the iterating values

```python
# print the number from 1 to 50
# range(initial_value, final_value + 1, step_size)

for i in range(1, 51,1):
    print(i)
```

### in the range there is never a neccessity for mentioning the step size and similarly if you want to print the numbers or you want to operate on numbers again there is no neccessity for passing the initial value as well if and only if your initial value is 0

```python
# numbers from 0 to 100
for i in range(101):
    print(i)
```

```python
# print numbers from 0 to 100 with gap of 5

for i in range(0,101,5):
    print(i)
```

## By using range() we can represent the numbers in decreamenting manner as well or in decreasing order as well

```python
# print the numbers from 50 to 1 with a gap of 1 value in a decreamenting manner

for i in range(50,0,-1):
    print(i)
```

```python
# print the numbers from 50 to 0 with a gap of 5

for i in range(50,-1,-5):
    print(i)
```

```python
# print numbers from 500 to 0 with a gap of 5

for i in range(500, -1, -5):
    print(i)
```
