### The Parent Exception type must be mentioned at the end or last as it is the parent class so only if the previous exception class will not catch the error then this final one will be able to catch it

### try---> try block is going to have the operation where we can except an error to occur 
### except---> which  is used to handle the error in a way such that the program execution shouldnt stop
### else--->it will be consisting of the code which we want to run if they try block operation is working without throwing any error
### finally--> Finally block gets executed once we are done with Everything and even if we are getting an error or not getting an error still the code inside the finally block gets executed

### if try---> code is throwing an Error ---> Except---> code will be running
### if try---> code is not throwing an Error--> Else---> code will be running
