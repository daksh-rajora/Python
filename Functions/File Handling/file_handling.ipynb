# File Handling
## 1. Read the content from My Text File in the system

```python
# open()--> opening your file
# first parameter---> file name
# second parameter---> mode
# Mode---> it decide the way in which we are working with this file
'''
Mode
r--> read mode--> default mode--> reading the content from the file
w--> write mode--> writing the content into the file
a--> append mode--> adding the content at the end of the file
r+--> read and write mode--> reading and writing the content into the file
w+--> write and read mode--> writing and reading the content from the file
a+--> append and read mode--> appending and reading the content from the file
'''
```

```python
# Reading file
# read()---> Reads the content from file

f=open("example.txt","r")
content=f.read()
print(content)
f.close()
```

## Writing the content into the file
## write()--> it overwrites the content into the text file
## append()---> it adds the data from the end of the file

```python
## write()---> helps us out in writing the content into the file
f=open("example.txt", "w")
f.write("This is my second line")
f.close()
```

```python
## appending the content into file
f=open('example.txt','a')
f.write('\n We are learning about gen ai ')
f.close()
```

## with---> it will be used for first of all opening the file as a object and there is no requirement for closing the file as well

```python
with open('example.txt','r') as f:
    content=f.read()
    print(content)
```

```python
with open('example.txt','w') as file:
    file.write('Hello World')
```

```python
with open('example.txt','a') as files:
    files.write('e\n Appending new content')
```
