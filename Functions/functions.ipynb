```python
# We are creating a Function to see that how the translation is going to be their from english to hindi

def translator(English_Text):
    """This function is going to be used for converting the english language into hindi language"""

    if English_Text=='Hello' or English_Text=='hello':
        print(f"The translation for {English_Text} is Namaste")

    if English_Text=='Help' or English_Text=='help':
        print(f"The translation for {English_Text} is Madad")

    if English_Text=='Required' or English_Text=='required':
        print(f"The translation for {English_Text} is Avashyak")
```

### For having the access to our Function and to provide the input to our function we can simply call the function

### Function calling is just like calling a waiter in a restaurent where in just like we call the waitor and provide the input as the food item and then waitor provide us with the food itself, like in a similar way we can call a function as well as well and can provide the input and can get the required output after calling the function only

```python
## calling a function --> Mention the name of the function and then provide the input

English_input=input("Enter the word that you want to get translated")

translator(English_input)
```

```python
## function to convert English into Hindi

def translator_return(English_text):

    if English_text=='Hello' or English_text=='hello':
        Hindi_text="Namaste"
    
    elif English_text=='Help' or English_text=='help':
        Hindi_text='Madad'

    elif English_text=='Required' or English_text=='required':
        Hindi_text='Avashyak'

    return Hindi_text
```

```python
English_text_input=input("Enter the text that you want to get translated")

Hindi_output=translator_return(English_text_input)
print(f"The translation of {English_text_input} is {Hindi_output}")
```

```python
# creating a function which is going to take input number from the user side and will figure out that
# whether the number is even or odd

def even_odd(number):
    if number%2==0:
        print(f"The number {number} is an even number")
    else:
        print(f"The number {number} is an odd number")
```

```python
even_odd(9)
```

```python
# Ask the user to provide with a proper input or if you are a good programmer then you wont
# be getting into this hassle of correcting the user and you can handle the issue on your own

import numpy as num
def even_or_odd(num):
    if isinstance(num,int):
        if num%2==0:
            print(f"The number {num} is an even number")
        else:
            print(f"The number {num} is an odd number")
    elif isinstance(num, str):
        print("Please enter a valid integer")
    else:
        print(f"{num} is not a valid integer")
```

```python
even_or_odd(5)
```

```python
even_or_odd('s')    
```

```python
even_or_odd(15.4)
```
