# Recursive Function

## By a recursive function we can say that a function is going to call itself again and again

```python
# Factorial of a Number
# Factorial of 8 --> 8*7*6*5*4*3*2*1
# Factorial of 98 --> 98*97*96....*1
# Factorial of 0 or factorial of 1 ---> 1
```

```python
def factorial(num):
    if num==0 or num==1:
        print(f'The factorial of {num} is 1')
    
    else:
        factorial=1
        for i in range(1,num+1):
            factorial=factorial*i
        print(f"The factorial of {num} is {factorial}")
```

```python
number=int(input("Enter the number whose factorial you want to calculate"))

factorial(number)
```

```python
def factorial_recursive(num):
    if num==0 or num==1:
        return 1

    else:
        return num*factorial_recursive(num-1)
```

```python
factorial_recursive(1)
```

```python
factorial_recursive(5)
```

```python
input_recursive=int(input("Enter the number whose factorial you want to calculate using recursion"))
print("The factorial of the given number is:", factorial_recursive(input_recursive))
```

## Function with variable length arguments

## Positional Arguments--> Position arguments are being used so that we can pass multiple values inside our function as an argument irrespective of the number of values we have mentioned as the parameter inside our function 

```python
# function to add multiple numbers irrespective of how many numbers we are passing

def addition_positional(*num):
    add=0
    for i in num:
        add+=i
    print('The addition of numbers are:',add)
```

```python
addition_positional(49,324,43,4)
```

## Keywords Arguments ---> Takes the input as the keyword values and irrespective of any number of keyword values you are going to pass it wont mater

```python
def data(**user_data):
    for key,value in user_data.items():
        print(f"The {key} is {value}")
```

```python
data(name='Max',age=20,gender='male')
```

## You gus want to pass keyword value pairs as your input along with positional arguments as well



```python
def variable(*positional, **keywords):
    for i in positional:
        print(f"The positional argument is: {i}")
    for key,value in keywords.items():
        print(f"key is: {key} and value is: {value}")
```

```python
variable(1,2,3,4,5,name='max',age=20,Qualification='Btech in computer science')
```
