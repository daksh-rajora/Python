# Filter Function

## Filter function is going to retieve only those items which are meeting certain condition in an iterable

## The filter function constructs an iterator from elements of an iterable for which a condition remains True or for which a Function returns True Value

## Syntax for Filter Function

### list(filter(func,iterable))

```python
# To retieve only those values which are even 

def even_function(value):
    if value%2==0:
        return True
```

```python
iterable_list=[1,23,32,454,62,3,6,7,4,5,7,67,]
print(list(filter(even_function, iterable_list)))
```

```python
# solving similar problem with lambda function
print(list(filter(lambda x: x%2==0, iterable_list)))
```

```python
# i want to retrieve the values which are greater than 5

print(list(filter(lambda x: x>5, iterable_list)))
```

```python
# i want to retrieve all those values which are greater than 5 as well and are even as well
list_items=[1,23,4,543,6,75,4,5,67,8]
print(list(filter(lambda x: x>5 and x%2==0, list_items)))
```

```python
# Iterating over the dictionary items

employees=[{'Name':'Max','Age':30,'Qualification':'Btech in Computer Science'},
           {'Name':'John','Age':25,'Qualification':'Mtech'},
           {'Name':'David','Age':32,'Qualification':'Btech in Electronics'}]
```

```python
def age_greater(employee):
    if employee['Age']>25:
        return True
```

```python
print(list(filter(age_greater, employees)))
```
