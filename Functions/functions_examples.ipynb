```python
# To create a function that converts temperature from celcius to Fahrenheit or F to Celsius

# if temperature is in celcius, then celcius to Fahrenheit conversion is

'''
1 degree celcius = 9/5 + 32
45 deg C = 45*9/5 + 32 F

1 Fahrenheit celcius = 5/9 - 32
45 deg F = 45-32 * 5/9 C
'''

def temperature_change(temp_value,unit):
    if unit == 'C':
        temperature=temp_value *9/5 + 32
        print(f'The temperature for {temp_value} deg celcius in Fahrenheit is :',str(temperature)+'F')

    elif unit == 'F':
        temperature= round((temp_value-32) * 5/9,2)
        print(f'The temperature for {temp_value} deg Fahrenheit in Celcius is :', str(temperature)+'C')

    else:
        print("Please enter valid tempertature unit")
```

```python
temperature_change(1,'C')
```

```python
temperature_change(1, 'F')
```

```python
# define a function that would behave as a Password Strength Checker

'''
Strong password -->
if your password:
    1.length is grater than 8 characters
    2. password must consist of at least a lowercase character (a,b,c....z)
    3. password must consist of at least a uppercase character (A, B, C....Z)
    4. password must consist of at least a digit (1,2,3...)
    5. password must consist of at least a special character (!, @, #, $, %, ^, &, *, (, ))

Weak Passwords-->
If a password doesnt match any above condition then it will be considered as a weak password
'''
```

## islower()--> checks that whether if the current character is a lowercase character or not

```python
def password_strength(password):
    if len(password)<8:
        print(f'The entered password {password} is a weak password')

    elif not any(char.islower() for char in password):
        print(f'The entered password {password} is a weak password')

    elif not any(char.isupper() for char in password):
        print(f'The entered password {password} is a weak password')

    elif not any(char.isdigit() for char in password):
        print(f'The entered password {password} is a weak password')

    elif not any(char in '!@#$%^&*()' for char in password):
        print(f'The entered password {password} is a weak password')

    elif len(password)>8 and any(char.islower() for char in password) and any(char.isupper() for char in password) and any(char.isdigit() for char in password) and any(char in '!@#$%^&*()' for char in password):
        print(f'The entered password {password} is a strong password')
```

# Palindrome String

### Palindrome string is going to have characters similar to it from both the sides or even if you will reverse the string still you will be getting the given string itself

### palindrome --> aba, aaabbaaa, zzyzz
### not palindrome --> Ab, abaa

## reverse string --> string[::-1]

```python
## Check that whether if the current string is a palindrome or not 
def palindrome(string):
    string_value=string.lower().replace(" ","")

    if string_value==string_value[::-1]:
        print(f'The entered string {string} is a palindrome')

    else:
        print(f'The entered string {string} is not a palindrome')
```

```python
palindrome('AB')
```

```python
palindrome('Aba')
```
