# Lambda Function

## Lambda functions are one liner functions defined by lambda keyword

## Syntax for Lambda Function:

## lambda arguments:expression

## lambda --> keyword which simply let the python interpreter know what the current function is our lambda function

## Arguments --> input values for our function

## Expression --> Logic(function Body)

#### lambda x:x**2

## They can have any number of arguments but only one expresion or only one logic
## The are used for short operations or as arguments to Higher Order Functions

## It can be treated as a function without a Name and they dont even require Return Statement

```python
# function which calculates the square of the numbers

def square(num):
    return num**2

```

```python
print(square(2))
```

```python
## This similar problem can be solved using lambda function as well

lambda_function=lambda x:x**2
```

```python
type(lambda_function)
```

```python
lambda_function(4)
```

```python
## lambda function to calculate sum of two numbers
lambda_sum=lambda x,y:x+y
print(lambda_sum(3,4))
```

```python
product=lambda a,x,y,z:a*x*y*z
print(product(1,2,3,4))
```
