# Map() function

### Map cuntion applies a given function operation to All items in an iterable or a list and returns a Map Object(An iterator)

### Mao function is particularly useful for Transforming the data or applying an operation on All items of the iterable

### Map() is going to help us out with avoiding unnecessary lines of code and we can apply the operation on all values or all item in an iterable and this is the sign of a good programmer

## Syntax for Map()

## list(map(function,iterable))

```python
# squaring all items of a list
# list--->[1,4,2,5,6,6,3]
# []--->[1,16,4,25,...]

def square(list_items):
    empty_list=[]
    for item in list_items:
        squared_items=item**2
        empty_list.append(squared_items)
    print(empty_list)
```

```python
square([1,2,34,5,5,])
```

```python
# similar problem with map() function
print(list(map(lambda x:x**2,[1,2,3,4,5])))
```

```python
# map function with basic function in it

def square_items(numbers):
    return numbers**2

print(list(map(square_items, [1, 2, 3, 4, 5, 6])))
```

## Whenever you are going to pass the general function inside your map function as the parameter then you dont have to use parenthesis

## While mapping the value of one iterable with another iterable we can actually apply the operation in between the values of the two iterable in a synchronous way
# [1,2,3,4] <---->[5,6,8,2]--->[6,8,11,6]

```python
# Map function with multiple iterables
number_1=[1,2,3,4]
number_2=[5,6,7,8]

print(list(map(lambda x,y:x+y,number_1,number_2)))
```

```python
# Map() to convert all string values items in an iterable into an integer

map_values=list(map(int, ['1', '2', '3', '4', '5']))
```

```python
print(map_values)
```

```python
employee_name=['amit','juhi','max','kunal','ansh']
print(list(map(str.upper,employee_name)))
```

```python
# Retrieve the values inside a dictionary using map()

employees=[{'name':'Amit','age':34,'designation':'SDE-1'},
           {'name':'Kunal','age':20,'designation':'SDE-2'},
           {'name':'Max','age':25,'designation':'SDE-3'}]
```

```python
def age_retrieve(employee):
    return employee['age']
```

```python
print(list(map(age_retrieve, employees)))
```
