# File Handling
## 1. Read the content from My Text File in the system

```python
# open()--> opening your file
# first parameter---> file name
# second parameter---> mode
# Mode---> it decide the way in which we are working with this file
'''
Mode
r--> read mode--> default mode--> reading the content from the file
w--> write mode--> writing the content into the file
a--> append mode--> adding the content at the end of the file
r+--> read and write mode--> reading and writing the content into the file
w+--> write and read mode--> writing and reading the content from the file
a+--> append and read mode--> appending and reading the content from the file
'''
```

```python
# Reading file
# read()---> Reads the content from file

f=open("example.txt","r")
content=f.read()
print(content)
f.close()
```

## Writing the content into the file
## write()--> it overwrites the content into the text file
## append()---> it adds the data from the end of the file

```python
## write()---> helps us out in writing the content into the file
f=open("example.txt", "w")
f.write("This is my second line")
f.close()
```

```python
## appending the content into file
f=open('example.txt','a')
f.write('\n We are learning about gen ai ')
f.close()
```

## with---> it will be used for first of all opening the file as a object and there is no requirement for closing the file as well

```python
with open('example.txt','r') as f:
    content=f.read()
    print(content)
```

```python
with open('example.txt','w') as file:
    file.write('Hello World')
```

```python
with open('example.txt','a') as files:
    files.write('e\n Appending new content')
```

```python
## Write  the content into file line by line ---> writelines()
lines=['First_Line\n','Second Line\n','Third Line\n']
with open('example.txt', 'w') as data:
    data.writelines(lines)
```

```python
# r , w , a
# Mode --> Write the connect into the binary file --> wb
# Mode --> Read the content from binary file --> rb

```

```python
data=b'\x00\x01\x02\x03\x04'
with open('example.txt','wb') as binary_data:
    binary_data.write(data)
```

```python
with open('example.txt', 'rb') as binary_read:
    binary_content=binary_read.read()
    print(binary_content)
```

```python
# Reading the Total lines, total words, total characters from the file

def count_lines(file):
    with open(file,'r') as files:
        lines=files.readlines()
        total_lines=len(lines)
        total_words=sum(len(line.split()) for line in lines)
        total_characters=sum(len(line) for line in lines)
    return total_lines, total_words, total_characters

file_path='Total_data.txt'
Total_lines, Total_Words, Total_Characters=count_lines(file_path)
print(f'Total Lines: {Total_lines} and Total Words: {Total_Words} and Total Characters: {Total_Characters}')
```

```python
## Write the content into a file and then after writing the content into that file we are going to read the content from that file
# as well ---> seek()
```

## w+ --> first of all W is their so it states that first writing operation will be done and then Reading operation will be done
## seek()--> it is going to help us out with reading the content after writing fron the starting only

```python
with open('Total_data.txt','w+') as write_read_file:
    write_read_file.write('\n Hello I am Love')
    write_read_file.write('\n I am 25 years old')
    write_read_file.seek(0) #--> brings the cursor to the starting position
    content=write_read_file.read()
    print(content)
```
