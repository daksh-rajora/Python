# OS Library

## OS library is able to interact with our operating system and can let us work with multiple directories and files

```python
## getcwd()---> it is capable of providing us with the idea about the current working directory
# getcwd()---> get current working directory
```

```python
import os
```

```python
current_path=os.getcwd()
print('The current working directory is :',current_path)
```

```python
## mkdir()--> make directory---> it will be helping is out with the creating the directory
```

```python
new_path='New_Folder'
os.mkdir(new_path)
print(f'The new directory {new_path} is created successfully')
```

```python
# listdir()--> Listing all of the directories present --> List Directories
All_Items=os.listdir('.')
print(All_Items)
```

```python
# Joining multiple Path together
# os.path.join()---> It will be joining multiple paths together
dir_name='Folder_Data'
file_name='File_Data'
full_path=os.path.join(dir_name,file_name)
print('The Full Path is :', full_path)
```

```python
# I want to joing multiple path with the current path

current_complete_path=os.path.join(os.getcwd(), dir_name, file_name)
print(f'The complete path is: {current_complete_path}')
```

```python
# os.path.exists()---> It will help us with checking whether a specific path exists or not
```

```python
# os.path.isfile()---> This checks the whether if the current path is a file or not
# os.path.isdir()---> This checks whether if the current path is a directory or not

path='Total_data.txt'

if os.path.isfile(path):
    print(f'The path {path} exists')
    if os.path.isfile(path):
        print(f'The path {path} is a File')
    elif os.path.isdir(path):
        print(f'The path {path} is a Directory')
```

```python
# The complete path<--> Absolute Path---> os.path.abspath()

# os.path.abspath()0---> It provides us with the path of the absolute path of the file or directory
```

```python
file_name='Total_data.txt' # Total_data.txt <---> Relative Path
complete_path=os.path.abspath(file_name)

print('The Absolute Path is :', complete_path)
```
