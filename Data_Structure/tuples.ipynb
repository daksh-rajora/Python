## Tuples


### Tuples are the second type of Data Structure which can store the data in a ordered way just like list but tuples are not mutable

### tuples will be created inside these brackets ()

```python
tuple_values=(1,2,3,4,5)
print(tuple_values)
```

### Tuples are like list only which can store the data inside it in a ordered way but are going to immutable

```python
tuple_values[1]=24
```

```python
list_values=[1,2,3,4,5]
list_values[1]
```

```python
list_values[1]=24
```

```python
list_values
```

### tuples are also capable of storing the data of any data type

```python
tuple_data=(1,2,3,4,5,"Daksh",True)
```

```python
tuple_data
```

```python
##Tuples are going to be ordered or are going to be indexed
tuple_data[:]
```

```python
tuple_data[1:]
```

```python
tuple_data[::-1]
```

```python
tuple_data[1:6]
```

```python
Nested_tuple[0][:]
```

```python
# Iterating over a Tuple inside a Tuple

for i in Nested_tuple:
    for j in i:
        print(j, sep="")
    print(" ")
```
