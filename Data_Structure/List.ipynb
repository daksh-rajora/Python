## List Methods
#### 1. append()--> insert the value at the end of the list
#### 2. insert()--> insert the value at a specific location inside the list
#### 3. remove()-->remove the value fron our list that we are going to pass inside it 
#### 4. pop()-->remove the value from the list at a specific index location
#### 5. index()-->it provide the index of the value that we are going to pass inside it if that value present inside the list
#### 6. sort()-->sort() is going to help us out with the sorting values in a ascending manner
#### 7. reverse()-->it can reverse the values present inside our list i.e, value at last index will be at first index now
#### 8. count()-->it counts the number of occurence of a specific value that we are going to pass inside our list and it returns that count of occurence
#### 9. clear()-->for removing all of the values that we are storing inside our list

## List Comprehension

## We are going to use looping statements inside the list so that we can store the values inside the list in an iterable format and in list comprehension conditional statements are allowedd as well

## Syntax for list comprehension is 
### [expression for item in iterable if condition]
### expression-->what we want to print
### for item in iterable-->looping statement
### if condition-->conditional statement

```python
#I need to store values from 1 to 50 inside the list
list_items=[x for x in range(1,51)] 
```

```python
list_items
```

```python
list_squared=[items**2 for items in range(1,16)]
```

```python
list_squared
```

```python
# if i am only interested in looking over the square values which are even
even_square=[items**2 for items in range(1, 16) if items**2%2==0]
```

```python
even_square
```

```python
##Functions in list comprehension as well
list_itemss = ["Daksh","Apple","banana","car","truck","laptop","technology"]
```

```python
length_list = [len(word) for word in list_itemss]
```

```python
length_list
```
